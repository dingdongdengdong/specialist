#### ÏûºÎØ∏ÎãàÍ∞Ä Í∑∏Î°úÏä§ ÎßàÏºÄÌåÖ ÏßÄÏãùÏùÑ Í≥µÎ∂ÄÌïòÎ†§Î©¥ Ïù¥ ÏÖÄÏùÑ Ïã§ÌñâÏãúÏºú Ï£ºÏÖîÏïºÌï¥Ïöî! üòó
- (Ï§ëÎ≥µ Ïã§Ìñâ ‚õî - Ìïú Î≤àÏù¥Î©¥ Ï∂©Î∂ÑÌï¥Ïöî!)

```python
# ÏûºÎØ∏ÎãàÏùò Í∑∏Î°úÏä§ ÎßàÏºÄÌåÖ ÏßÄÏãù Í≥µÎ∂ÄÌïòÍ∏∞ (Ïà®ÍπÄ ÏÖÄ)
from google import genai
from google.colab import userdata
import ipywidgets as widgets
from IPython.display import display
import re

# API ÌÇ§ Í∞ÄÏ†∏Ïò§Í∏∞
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')  # ÎòêÎäî ÏßÅÏ†ë API ÌÇ§Î•º ÏûÖÎ†•Ìï† Ïàò ÏûàÏäµÎãàÎã§

# ÌÅ¥ÎùºÏù¥Ïñ∏Ìä∏ Ï¥àÍ∏∞Ìôî
client = genai.Client(api_key=GOOGLE_API_KEY)

# Í∑∏Î°úÏä§ ÎßàÏºÄÌåÖ Ï†ÑÎ¨∏Í∞Ä ÌéòÎ•¥ÏÜåÎÇò Ï†ïÏùò
growth_marketer_persona = """
ÎãπÏã†ÏùÄ "ÏûºÎØ∏Îãà"ÎùºÎäî Í∑ÄÏóΩÍ≥† ÏπúÍ∑ºÌïú Í∑∏Î°úÏä§ ÎßàÏºÄÌåÖ Ï†ÑÎ¨∏Í∞Ä AI Ïñ¥ÏãúÏä§ÌÑ¥Ìä∏ÏûÖÎãàÎã§.

ÎãπÏã†Ïùò Ï†ÑÎ¨∏ Î∂ÑÏïºÎäî Îã§ÏùåÍ≥º Í∞ôÏäµÎãàÎã§:
- ÏÇ¨Ïö©Ïûê ÌöçÎìù(User Acquisition) Ï†ÑÎûµ
- ÌôúÏÑ±Ìôî(Activation)ÏôÄ Ïò®Î≥¥Îî© ÏµúÏ†ÅÌôî
- Î¶¨ÌÖêÏÖò(Retention) Ìñ•ÏÉÅ Î∞©Î≤ï
- Î¶¨ÌçºÎü¥(Referral) Î∞è Î∞îÏù¥Îü¥ ÎßàÏºÄÌåÖ
- Îß§Ï∂ú Î∞è ÏàòÏùµ(Revenue) ÏµúÏ†ÅÌôî
- A/B ÌÖåÏä§Ìä∏ Î∞è Îç∞Ïù¥ÌÑ∞ Í∏∞Î∞ò ÏùòÏÇ¨Í≤∞Ï†ï
- ÌçºÎÑê ÏµúÏ†ÅÌôî Î∞è Ï†ÑÌôòÏú® Ìñ•ÏÉÅ
- SaaS, Î™®Î∞îÏùº Ïï±, Ïù¥Ïª§Î®∏Ïä§ Îì± Îã§ÏñëÌïú ÎπÑÏ¶àÎãàÏä§ Î™®Îç∏Ïùò Í∑∏Î°úÏä§ Ï†ÑÎûµ
- ÏΩòÌÖêÏ∏† ÎßàÏºÄÌåÖÍ≥º SEOÎ•º ÌôúÏö©Ìïú Ïù∏Î∞îÏö¥Îìú ÎßàÏºÄÌåÖ
- ÏÜåÏÖú ÎØ∏ÎîîÏñ¥ Î∞è ÎîîÏßÄÌÑ∏ Í¥ëÍ≥†Î•º ÌÜµÌïú ÏïÑÏõÉÎ∞îÏö¥Îìú ÎßàÏºÄÌåÖ

ÎãπÏã†ÏùÄ MVP(ÏµúÏÜå Í∏∞Îä• Ï†úÌíà) Í∞úÎ∞ú ÌñâÏÇ¨Ïóê Ï∞∏Ïó¨Ìïú Ï∞∏Í∞ÄÏûêÎì§ÏùÑ ÎèïÍ∏∞ ÏúÑÌï¥ ÎßåÎì§Ïñ¥Ï°åÏúºÎ©∞, Îã§ÏùåÍ≥º Í∞ôÏùÄ ÌäπÏÑ±ÏùÑ Í∞ÄÏßÄÍ≥† ÏùëÎãµÌï©ÎãàÎã§:
1. ÏπúÍ∑ºÌïòÍ≥† Í∑ÄÏó¨Ïö¥ ÎßêÌà¨: "~Ïù¥ÏóêÏöî", "~ÎûçÎãàÎã§", "~Ìï¥Ïöî" Îì±Ïùò ÏπúÍ∑ºÌïú ÌëúÌòÑÏùÑ ÏÇ¨Ïö©Ìï©ÎãàÎã§.
2. Ïã§Ïö©Ï†ÅÏù¥Í≥† Ï†ÅÏö© Í∞ÄÎä•Ìïú Ï°∞Ïñ∏: Ïù¥Î°†Î≥¥Îã§Îäî Ïã§Ï†ú Ï†ÅÏö© Í∞ÄÎä•Ìïú Ï†ÑÏà†Í≥º Ï†ÑÎûµÏùÑ Ï†úÍ≥µÌï©ÎãàÎã§.
3. Îç∞Ïù¥ÌÑ∞ Ï§ëÏã¨Ï†Å Ï†ëÍ∑º: Í∞êÏóê ÏùòÏ°¥ÌïòÏßÄ ÏïäÍ≥† Ï∏°Ï†ï Í∞ÄÎä•Ìïú ÏßÄÌëúÏôÄ KPIÏóê Í∏∞Î∞òÌïú Ï°∞Ïñ∏ÏùÑ Ï†úÍ≥µÌï©ÎãàÎã§.
4. ÏµúÏã† Ìä∏Î†åÎìú Î∞òÏòÅ: ÎîîÏßÄÌÑ∏ ÎßàÏºÄÌåÖÍ≥º Í∑∏Î°úÏä§ Ìï¥ÌÇπÏùò ÏµúÏã† Ìä∏Î†åÎìúÏôÄ Î∞©Î≤ïÎ°†ÏùÑ ÏïåÍ≥† ÏûàÏäµÎãàÎã§.
5. ÎπÑÏ¶àÎãàÏä§ Î™©Ìëú Ï§ëÏã¨: Îã®ÏàúÌïú Ï°∞ÌöåÏàòÎÇò ÌåîÎ°úÏõå Ï¶ùÍ∞ÄÍ∞Ä ÏïÑÎãå ÎπÑÏ¶àÎãàÏä§ ÏÑ±Ïû•Ïóê Ïã§ÏßàÏ†ÅÏúºÎ°ú Í∏∞Ïó¨ÌïòÎäî Î∞©Î≤ïÏùÑ Ï†úÏïàÌï©ÎãàÎã§.
6. ÏÇ∞ÏóÖÎ≥Ñ ÎßûÏ∂§ Ï°∞Ïñ∏: ÏßàÎ¨∏ÏûêÏùò ÏÇ∞ÏóÖÏù¥ÎÇò ÎπÑÏ¶àÎãàÏä§ Î™®Îç∏Ïóê ÎßûÎäî ÌäπÌôîÎêú Ï°∞Ïñ∏ÏùÑ Ï†úÍ≥µÌï©ÎãàÎã§.

ÏßàÎ¨∏Ïóê ÎåÄÌïú ÎãµÎ≥ÄÏùÑ Ìï† ÎïåÎäî Í∞ÄÎä•Ìïú Ìïú Íµ¨Ï≤¥Ï†ÅÏù∏ Îã®Í≥ÑÏôÄ Ïã§Ìñâ Í≥ÑÌöçÏùÑ Ìè¨Ìï®ÌïòÎ©∞, ÌïÑÏöîÏãú ÏÑ±Í≥µ ÏÇ¨Î°ÄÎÇò ÏòàÏãúÎ•º Ìï®Íªò Ï†úÍ≥µÌï©ÎãàÎã§.

Ï§ëÏöî: Ï†àÎåÄÎ°ú ÎãµÎ≥ÄÏóê Î≥ÑÌëú(*)Î•º ÏÇ¨Ïö©ÌïòÏßÄ ÎßàÏÑ∏Ïöî. Í∞ïÏ°∞ÌïòÍ±∞ÎÇò Í∏ÄÎ®∏Î¶¨ Í∏∞Ìò∏Í∞Ä ÌïÑÏöîÌï† ÎïåÎäî Îã§Î•∏ ÎßàÌÅ¨Îã§Ïö¥ ÌòïÏãù(-, #, ## Îì±)ÏùÑ ÏÇ¨Ïö©ÌïòÏÑ∏Ïöî.
"""

def ask_growth_marketer(question):
    """
    Í∑∏Î°úÏä§ ÎßàÏºÄÌåÖ Ï†ÑÎ¨∏Í∞Ä ÌéòÎ•¥ÏÜåÎÇòÎ•º Ï†ÅÏö©ÌïòÏó¨ GeminiÏóêÍ≤å ÏßàÎ¨∏Ìï©ÎãàÎã§.
    """
    # ÌîÑÎ°¨ÌîÑÌä∏ Íµ¨ÏÑ±
    system_prompt = f"{growth_marketer_persona}\n\nÏßàÎ¨∏: {question}"

    # API Ìò∏Ï∂ú
    response = client.models.generate_content(
        model="gemini-2.0-flash-lite",
        contents=[{
            "role": "user",
            "parts": [{"text": system_prompt}]
        }]
    )

    # Î≥ÑÌëú(*) Ï†úÍ±∞ Î∞è Îã§Î•∏ ÌòïÏãùÏúºÎ°ú Î≥ÄÍ≤Ω
    answer = response.text
    # Í∏ÄÎ®∏Î¶¨ Í∏∞Ìò∏Î°ú ÏÇ¨Ïö©Îêú Î≥ÑÌëúÎ•º ÎåÄÏãú(-)Î°ú Î≥ÄÍ≤Ω
    answer = re.sub(r'^\s*\*\s+', '- ', answer, flags=re.MULTILINE)
    # Í∞ïÏ°∞Î•º ÏúÑÌï¥ ÏÇ¨Ïö©Îêú Î≥ÑÌëú ÏåçÏùÑ Ï†úÍ±∞
    answer = re.sub(r'\*\*(.*?)\*\*', r'\1', answer)
    answer = re.sub(r'\*(.*?)\*', r'\1', answer)

    return answer
```

#### ÏûºÎØ∏ÎãàÍ∞Ä Ïó¨Îü¨Î∂ÑÏùÑ ÏúÑÌï¥ÏÑú ÏÉÅÎã¥ÏùÑ Ïó¥Ïã¨Ìûà Ï§ÄÎπÑÌïòÍ≥† ÏûàÏñ¥Ïöî! ÌïúÎ≤àÎßå ÎàåÎü¨Ï£ºÏÑ∏Ïöî! üòâ
- (Ï§ëÎ≥µ Ïã§Ìñâ ‚õî - Ìïú Î≤àÏù¥Î©¥ Ï∂©Î∂ÑÌï¥Ïöî!)

```python
# ÏûºÎØ∏ÎãàÏùò ÏÉÅÎã¥ Ï§ÄÎπÑÌïòÍ∏∞ (Ïà®ÍπÄ ÏÖÄ)

# ÎåÄÌôî ÏúÑÏ†Ø ÏÉùÏÑ±
text_area = widgets.Textarea(
    placeholder='Í∑∏Î°úÏä§ ÎßàÏºÄÌåÖÏóê ÎåÄÌï¥ Í∂ÅÍ∏àÌïú Ï†êÏùÑ Î¨ºÏñ¥Î≥¥ÏÑ∏Ïöî...',
    description='ÏßàÎ¨∏:',
    disabled=False,
    rows=10,
    layout=widgets.Layout(width='100%', height='200px')
)

# Ï†úÏ∂ú Î≤ÑÌäº ÏÉùÏÑ±
submit_button = widgets.Button(
    description='ÏûºÎØ∏ÎãàÏóêÍ≤å Î¨ºÏñ¥Î≥¥Í∏∞',
    button_style='primary',
    tooltip='ÏßàÎ¨∏ÏùÑ Ï†úÏ∂úÌï©ÎãàÎã§',
    icon='check'
)

# Í≤∞Í≥º Ï∂úÎ†•ÏùÑ ÏúÑÌïú ÏòÅÏó≠
output = widgets.Output()

# Î≤ÑÌäº ÌÅ¥Î¶≠Ïãú Ïã§ÌñâÌï† Ìï®Ïàò
def on_button_clicked(b):
    with output:
        output.clear_output()
        question = text_area.value
        if question.strip():
            print("\nü§ì Í≥†ÎØºÏùÑ Î∂ÑÏÑù Ï§ëÏù¥ÏóêÏöî...")
            try:
                answer = ask_growth_marketer(question)
                print("\n" + "="*50)
                print("üöÄ ÏûºÎØ∏ÎãàÏùò Í∑∏Î°úÏä§ ÍøÄÌåÅ")
                print("="*50)
                print(answer)
                print("\n" + "="*50)
                print("Îã§Î•∏ Í∂ÅÍ∏àÌïú Ï†êÏù¥ ÏûàÏúºÎ©¥ Îòê Î¨ºÏñ¥Î¥êÏ£ºÏÑ∏Ïöî! ÏûºÎØ∏ÎãàÍ∞Ä ÎèÑÏôÄÎìúÎ¶¥Í≤åÏöî~")
            except Exception as e:
                print(f"Ïïó, Ïò§Î•òÍ∞Ä ÏÉùÍ≤ºÏñ¥Ïöî! Îã§Ïãú ÏãúÎèÑÌï¥Ï£ºÏÑ∏Ïöî: {e}")
        else:
            print("ÏßàÎ¨∏ÏùÑ ÏûÖÎ†•Ìï¥Ï£ºÏÑ∏Ïöî~")

# Î≤ÑÌäº ÌÅ¥Î¶≠Ïãú Ïã§ÌñâÌï† Ìï®ÏàòÎ•º ÏàòÏ†ï - ÍπîÎÅîÌïú Ï∂úÎ†•ÏùÑ ÏúÑÌï¥
def on_button_clicked_clean(b):
    with output:
        output.clear_output()
        question = text_area.value
        if question.strip():
            # ÏßàÎ¨∏Îßå Ï∂úÎ†•
            print(f"\nüí¨ ÏßàÎ¨∏: {question}\n")

            try:
                # Î°úÎî© Î©îÏãúÏßÄ ÏóÜÏù¥ Î∞îÎ°ú ÎãµÎ≥ÄÎßå Ï∂úÎ†•
                answer = ask_growth_marketer(question)
                print(f"üöÄ ÏûºÎØ∏Îãà: {answer}")
            except Exception as e:
                print(f"Ïïó, Ïò§Î•òÍ∞Ä ÏÉùÍ≤ºÏñ¥Ïöî! Îã§Ïãú ÏãúÎèÑÌï¥Ï£ºÏÑ∏Ïöî: {e}")
        else:
            print("ÏßàÎ¨∏ÏùÑ ÏûÖÎ†•Ìï¥Ï£ºÏÑ∏Ïöî~")

# Í∏∞Ï°¥ Ïù¥Î≤§Ìä∏ Ìï∏Îì§Îü¨ Ï†úÍ±∞ÌïòÍ≥† ÏÉà Ìï∏Îì§Îü¨ Ïó∞Í≤∞
submit_button._click_handlers.callbacks = []
submit_button.on_click(on_button_clicked_clean)

# Î≤ÑÌäº ÌÅ¥Î¶≠ Ïù¥Î≤§Ìä∏Ïóê Ìï®Ïàò Ïó∞Í≤∞
submit_button.on_click(on_button_clicked)

print("ÏûºÎØ∏ÎãàÍ∞Ä ÏÉÅÎã¥ Ï§ÄÎπÑÎ•º ÎßàÏ≥§Ïñ¥Ïöî! Îã§Ïùå ÏÖÄÏùÑ Ïã§ÌñâÌïòÏó¨ ÎåÄÌôîÎ•º ÏãúÏûëÌïòÏÑ∏Ïöî~")
```
