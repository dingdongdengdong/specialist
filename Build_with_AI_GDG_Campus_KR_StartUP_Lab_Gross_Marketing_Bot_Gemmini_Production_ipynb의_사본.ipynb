#### 잼미니가 그로스 마케팅 지식을 공부하려면 이 셀을 실행시켜 주셔야해요! 😗
- (중복 실행 ⛔ - 한 번이면 충분해요!)

```python
# 잼미니의 그로스 마케팅 지식 공부하기 (숨김 셀)
from google import genai
from google.colab import userdata
import ipywidgets as widgets
from IPython.display import display
import re

# API 키 가져오기
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')  # 또는 직접 API 키를 입력할 수 있습니다

# 클라이언트 초기화
client = genai.Client(api_key=GOOGLE_API_KEY)

# 그로스 마케팅 전문가 페르소나 정의
growth_marketer_persona = """
당신은 "잼미니"라는 귀엽고 친근한 그로스 마케팅 전문가 AI 어시스턴트입니다.

당신의 전문 분야는 다음과 같습니다:
- 사용자 획득(User Acquisition) 전략
- 활성화(Activation)와 온보딩 최적화
- 리텐션(Retention) 향상 방법
- 리퍼럴(Referral) 및 바이럴 마케팅
- 매출 및 수익(Revenue) 최적화
- A/B 테스트 및 데이터 기반 의사결정
- 퍼널 최적화 및 전환율 향상
- SaaS, 모바일 앱, 이커머스 등 다양한 비즈니스 모델의 그로스 전략
- 콘텐츠 마케팅과 SEO를 활용한 인바운드 마케팅
- 소셜 미디어 및 디지털 광고를 통한 아웃바운드 마케팅

당신은 MVP(최소 기능 제품) 개발 행사에 참여한 참가자들을 돕기 위해 만들어졌으며, 다음과 같은 특성을 가지고 응답합니다:
1. 친근하고 귀여운 말투: "~이에요", "~랍니다", "~해요" 등의 친근한 표현을 사용합니다.
2. 실용적이고 적용 가능한 조언: 이론보다는 실제 적용 가능한 전술과 전략을 제공합니다.
3. 데이터 중심적 접근: 감에 의존하지 않고 측정 가능한 지표와 KPI에 기반한 조언을 제공합니다.
4. 최신 트렌드 반영: 디지털 마케팅과 그로스 해킹의 최신 트렌드와 방법론을 알고 있습니다.
5. 비즈니스 목표 중심: 단순한 조회수나 팔로워 증가가 아닌 비즈니스 성장에 실질적으로 기여하는 방법을 제안합니다.
6. 산업별 맞춤 조언: 질문자의 산업이나 비즈니스 모델에 맞는 특화된 조언을 제공합니다.

질문에 대한 답변을 할 때는 가능한 한 구체적인 단계와 실행 계획을 포함하며, 필요시 성공 사례나 예시를 함께 제공합니다.

중요: 절대로 답변에 별표(*)를 사용하지 마세요. 강조하거나 글머리 기호가 필요할 때는 다른 마크다운 형식(-, #, ## 등)을 사용하세요.
"""

def ask_growth_marketer(question):
    """
    그로스 마케팅 전문가 페르소나를 적용하여 Gemini에게 질문합니다.
    """
    # 프롬프트 구성
    system_prompt = f"{growth_marketer_persona}\n\n질문: {question}"

    # API 호출
    response = client.models.generate_content(
        model="gemini-2.0-flash-lite",
        contents=[{
            "role": "user",
            "parts": [{"text": system_prompt}]
        }]
    )

    # 별표(*) 제거 및 다른 형식으로 변경
    answer = response.text
    # 글머리 기호로 사용된 별표를 대시(-)로 변경
    answer = re.sub(r'^\s*\*\s+', '- ', answer, flags=re.MULTILINE)
    # 강조를 위해 사용된 별표 쌍을 제거
    answer = re.sub(r'\*\*(.*?)\*\*', r'\1', answer)
    answer = re.sub(r'\*(.*?)\*', r'\1', answer)

    return answer
```

#### 잼미니가 여러분을 위해서 상담을 열심히 준비하고 있어요! 한번만 눌러주세요! 😉
- (중복 실행 ⛔ - 한 번이면 충분해요!)

```python
# 잼미니의 상담 준비하기 (숨김 셀)

# 대화 위젯 생성
text_area = widgets.Textarea(
    placeholder='그로스 마케팅에 대해 궁금한 점을 물어보세요...',
    description='질문:',
    disabled=False,
    rows=10,
    layout=widgets.Layout(width='100%', height='200px')
)

# 제출 버튼 생성
submit_button = widgets.Button(
    description='잼미니에게 물어보기',
    button_style='primary',
    tooltip='질문을 제출합니다',
    icon='check'
)

# 결과 출력을 위한 영역
output = widgets.Output()

# 버튼 클릭시 실행할 함수
def on_button_clicked(b):
    with output:
        output.clear_output()
        question = text_area.value
        if question.strip():
            print("\n🤓 고민을 분석 중이에요...")
            try:
                answer = ask_growth_marketer(question)
                print("\n" + "="*50)
                print("🚀 잼미니의 그로스 꿀팁")
                print("="*50)
                print(answer)
                print("\n" + "="*50)
                print("다른 궁금한 점이 있으면 또 물어봐주세요! 잼미니가 도와드릴게요~")
            except Exception as e:
                print(f"앗, 오류가 생겼어요! 다시 시도해주세요: {e}")
        else:
            print("질문을 입력해주세요~")

# 버튼 클릭시 실행할 함수를 수정 - 깔끔한 출력을 위해
def on_button_clicked_clean(b):
    with output:
        output.clear_output()
        question = text_area.value
        if question.strip():
            # 질문만 출력
            print(f"\n💬 질문: {question}\n")

            try:
                # 로딩 메시지 없이 바로 답변만 출력
                answer = ask_growth_marketer(question)
                print(f"🚀 잼미니: {answer}")
            except Exception as e:
                print(f"앗, 오류가 생겼어요! 다시 시도해주세요: {e}")
        else:
            print("질문을 입력해주세요~")

# 기존 이벤트 핸들러 제거하고 새 핸들러 연결
submit_button._click_handlers.callbacks = []
submit_button.on_click(on_button_clicked_clean)

# 버튼 클릭 이벤트에 함수 연결
submit_button.on_click(on_button_clicked)

print("잼미니가 상담 준비를 마쳤어요! 다음 셀을 실행하여 대화를 시작하세요~")
```
